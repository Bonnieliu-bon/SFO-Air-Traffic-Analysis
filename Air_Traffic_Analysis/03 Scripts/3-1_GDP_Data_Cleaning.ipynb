```python
duplicated_rows = df[df.duplicated(keep=False)]
duplicated_rows.head()
```

```python
df_dropped_rows = df.copy()
# Define the columns to clean and convert
cols_to_clean = [str(year) for year in range(2007, 2016)]

# First, handle the string '..' by replacing it with a NaN value
df_dropped_rows.replace('..', np.nan, inplace=True)

# Remove commas and convert the specified columns to numeric type
for col in cols_to_clean:
    df_dropped_rows[col] = df_dropped_rows[col].str.replace(',', '', regex=False).astype(float)
```

```python
df_dropped_rows.dtypes
```

```python
# addressing mixed-type data in a column

for col in df_dropped_rows.columns.tolist():
    # Get the type of each element in the column
    types_in_column = df_dropped_rows[col].map(type)
    
    # Get the type of the first element in the column
    # We use type() directly on the element, not .apply(type)
    type_of_first_element = type(df_dropped_rows[col].iloc[0])
    
    # Check if any element's type is different from the first element's type
    weird = (types_in_column != type_of_first_element).any()
    
    if weird:
        print(f"Column '{col}' contains mixed data types.")

print("Finished checking for mixed data types.")
```

```python
# Identify the columns that should be numeric (your year columns)
numeric_cols = [str(year) for year in range(2007, 2016)]

# Apply pd.to_numeric with errors='coerce' to each of these columns
for col in numeric_cols:
    df_dropped_rows[col] = pd.to_numeric(df_dropped_rows[col], errors='coerce')

print("Data types after coercing to numeric:")
print(df_dropped_rows.dtypes)

print("\nFirst few rows after coercing (check for NaNs):")
print(df_dropped_rows.head().to_markdown(index=False, numalign='left', stralign='left'))

print("\nNumber of NaN values per numeric column after coercing:")
print(df_dropped_rows[numeric_cols].isnull().sum())
```

```python
rows_with_nan = df_dropped_rows[df_dropped_rows.isnull().any(axis=1)]

print(f"Number of rows containing NaN values: {rows_with_nan.shape[0]}\n")
print("Rows with NaN values:")
print(rows_with_nan.to_markdown(index=False, numalign='left', stralign='left'))
```

```python
rows_with_nan.shape
```

```python
#Drop rows with any NaN in the numeric columns

df_cleaned_dropped = df_dropped_rows.dropna(subset=numeric_cols)
print("\nShape after dropping rows with NaNs:", df_cleaned_dropped.shape)
```
